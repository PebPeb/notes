```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10, (10*100) + 1)
```

```python
## ----------------------------------- ##
## ReLU
## ----------------------------------- ##

def ReLU(x):
  return [y if y >= 0 else 0 for y in x]
y = ReLU(x)

# Plotting ReLU
plt.figure(figsize=(6,3))
plt.plot(x, y)

plt.grid(True)
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.xticks(range(-10, 11, 2))

plt.show()
```

```python
## ----------------------------------- ##
## Sigmoid
## ----------------------------------- ##

def sigmoid(x):
  return [1 / (1 + np.exp(-y)) for y in x]

y = sigmoid(x)

# Plotting ReLU
plt.figure(figsize=(6,3))
plt.plot(x, y)

plt.grid(True)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.xticks(range(-10, 11, 2))

plt.show()
```

```python
## ----------------------------------- ##
## Tanh
## ----------------------------------- ##

def Tanh(x):
  return [(1 - np.exp(-2*y)) / (1 + np.exp(-2*y)) for y in x]

y = Tanh(x)

# Plotting ReLU
plt.figure(figsize=(6,3))
plt.plot(x, y)

plt.grid(True)
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.xticks(range(-10, 11, 2))

plt.show()

```
